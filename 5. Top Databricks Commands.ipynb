## Python / PySpark (notebook)

```python
df = spark.read.csv("/mnt/data/file.csv", header=True, inferSchema=True)
df.show()
df.printSchema()
```

```python
df.write.mode("overwrite").parquet("/mnt/data/output/")
```

```python
df.select("column1", "column2")
df.filter(df["column1"] > 100)
df.groupBy("column").agg({"amount": "sum"})
```

```python
df.cache()
df.unpersist()
```

```python
display(df)
```

## SQL

```python
%sql
SELECT * FROM my_table;
SELECT name, count(*) FROM my_table GROUP BY name;
```

```python
%sql
CREATE TABLE IF NOT EXISTS my_table (id INT, name STRING);
DROP TABLE IF EXISTS my_table;
```

```python
%sql
INSERT INTO my_table VALUES (1, 'Alice');
```

```python
%sql
DESCRIBE TABLE my_table;
SHOW TABLES IN my_database;
SHOW COLUMNS IN my_table;
```
